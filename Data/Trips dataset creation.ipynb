### Add the weather attributes

```python
import pandas as pd

# Define the file paths for your CSV files
#combined_lyft_path = r'C:\Users\olivi\OneDrive\Documents\McGill MMA\Fall 2023\Data mining (INSY662)\Project\Lyft data\combined_lyft.csv'
weather_path = r'C:\Users\olivi\OneDrive\Documents\McGill MMA\Fall 2023\Data mining (INSY662)\Project\weather.csv'

# Read the CSV files into dataframes
#combined_lyft = pd.read_csv(combined_lyft_path)
weather = pd.read_csv(weather_path)

# Perform a left join on the "start_time" column in combined_data and the "date" column in weather
lyft_weather = pd.merge(combined_lyft, weather, left_on=combined_lyft['started_at'].str[:10], right_on=weather['date'], how='left')

# Drop the redundant 'date' column that comes from the 'weather' dataset
lyft_weather.drop(columns='date', inplace=True)

# If it created a 'key_0' column : drop it
lyft_weather.drop(columns='key_0', inplace=True)
```

```python
import holidays

# Create a list of US holidays
us_holidays = holidays.US()

# New Year's Day (January 1)
# Martin Luther King Jr. Day (third Monday in January)
# Presidents' Day (third Monday in February)
# Memorial Day (last Monday in May)
# Independence Day (July 4)
# Labor Day (first Monday in September)
# Columbus Day (second Monday in October)
# Veterans Day (November 11)
# Thanksgiving Day (fourth Thursday in November)
# Christmas Day (December 25)

# Define a function to check if a date is a US holiday
def is_us_holiday(date):
    return date in us_holidays

# Apply the function to create the "Holiday" column
lyft_weather['Holiday'] = lyft_weather['started_at'].apply(is_us_holiday).astype(int)

lyft_weather_holiday = lyft_weather

# Now, lyft_weather['Holiday'] will contain binary values (1 for US holidays and 0 for non-holidays)
```

### Add time elapsed between start and end

```python
# Convert the 'started_at' and 'ended_at' columns to datetime objects
lyft_weather_holiday['started_at'] = pd.to_datetime(lyft_weather_holiday['started_at'])
lyft_weather_holiday['ended_at'] = pd.to_datetime(lyft_weather_holiday['ended_at'])

# Calculate the time difference in minutes and store it in a new column 'trip_time'
lyft_weather_holiday['trip_time'] = (lyft_weather_holiday['ended_at'] - lyft_weather_holiday['started_at']).dt.total_seconds() / 60
```

### Add distance traveled between start and end

```python
import pandas as pd
import numpy as np

# Define a function to calculate the Haversine distance between two sets of coordinates
def haversine_distance(lat1, lon1, lat2, lon2):
    # Radius of the Earth in kilometers
    R = 6371.0

    # Convert latitude and longitude from degrees to radians
    lat1, lon1, lat2, lon2 = np.radians([lat1, lon1, lat2, lon2])

    # Haversine formula
    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = np.sin(dlat / 2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon / 2)**2
    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1 - a))

    # Calculate the distance
    distance = R * c
    return distance

# Assuming your DataFrame is named lyft_weather_holiday

# Calculate the distance traveled using the Haversine formula and store it in a new column 'distance_traveled'
lyft_weather_holiday['distance_traveled'] = lyft_weather_holiday.apply(
    lambda row: haversine_distance(row['start_lat'], row['start_lng'], row['end_lat'], row['end_lng']), axis=1)
```
