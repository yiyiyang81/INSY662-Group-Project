```python
import pandas as pd

path='C:\\Users\\olivi\\OneDrive\\Documents\\McGill MMA\\Fall 2023\\Data mining (INSY662)\\Project\\Lyft data\\'
df = pd.read_csv(path+"complete_dataset_V2.csv", sep=',')

df.head(5)
```

```python
import pandas as pd

# Assuming your original DataFrame is named 'df'
# Replace these column names with the actual column names from your dataset
columns_to_select = ['started_at', 'start_station_cluster', 'rideable_type']

# Select relevant columns
selected_data = df[columns_to_select]

# Convert 'started_at' to datetime format if it's not already
selected_data['started_at'] = pd.to_datetime(selected_data['started_at'])

# Replace 'docked_bike' with 'classic_bike' in the 'rideable_type' column
selected_data['rideable_type'] = selected_data['rideable_type'].replace({'docked_bike': 'classic_bike'})

# Create a new column for the date without time
selected_data['date'] = selected_data['started_at'].dt.date

# Group by 'date', 'start_station_cluster', and 'rideable_type', then count the number of trips
demand_df = selected_data.groupby(['date', 'start_station_cluster', 'rideable_type']).size().reset_index(name='trip_count')

# Display the resulting DataFrame
demand_df
```

```python
# Save
path = r'C:\Users\olivi\OneDrive\Documents\McGill MMA\Fall 2023\Data mining (INSY662)\Project\Lyft data\dataset_demand.csv'
demand_df.to_csv(path, index=False)
```
