```python
import pandas as pd

path='C:\\Users\\olivi\\OneDrive\\Documents\\McGill MMA\\Fall 2023\\Data mining (INSY662)\\Project\\Lyft data\\'
df = pd.read_csv(path+"complete_dataset.csv", sep=',')

df.head(5)
```

### Add months

```python
import calendar

# Convert the 'started_at' column to datetime, if it's not already
df['started_at'] = pd.to_datetime(df['started_at'])

# Create a new 'month' column by mapping the month number to month names
df['month'] = df['started_at'].dt.month.map(lambda x: calendar.month_name[x])

df.head(5)
```

### Add hours

```python
df['started_at'] = pd.to_datetime(df['started_at'])
df['hour'] = df['started_at'].dt.hour

df.head(5)
```

### Add wind direction categories

```python
# Define a function to categorize wind direction into four cardinal directions
def categorize_wind_direction(degrees):
    if (315 <= degrees < 360) or (0 <= degrees < 45):
        return 'North'
    elif 45 <= degrees < 135:
        return 'East'
    elif 135 <= degrees < 225:
        return 'South'
    elif 225 <= degrees < 315:
        return 'West'

# Create a new column 'wind_direction_category' based on the 'wdir' column
df['wind_direction'] = df['wdir'].apply(categorize_wind_direction)

df.head(5)
```

### Clustering the stations

```python
#start

# Select the relevant columns for stations
station_columns = ['start_station_id', 'start_station_name', 'start_lat', 'start_lng']

# Drop duplicates and keep the first occurrence
start_stations_df = df[station_columns].drop_duplicates(subset=['start_station_id'])

# Rename the columns for consistency
start_stations_df = start_stations_df.rename(columns={
    'start_station_id': 'station_id',
    'start_station_name': 'station_name',
    'start_lat': 'station_lat',
    'start_lng': 'station_lng'
})

# Reset the index to have consecutive integer indices
start_stations_df = start_stations_df.reset_index(drop=True)

# Display the resulting DataFrame
start_stations_df

################################################################

#end
# Select the relevant columns for stations
station_columns = ['end_station_id', 'end_station_name', 'end_lat', 'end_lng']

# Drop duplicates and keep the first occurrence
end_stations_df = df[station_columns].drop_duplicates(subset=['end_station_id'])

# Rename the columns for consistency
end_stations_df = end_stations_df.rename(columns={
    'end_station_id': 'station_id',
    'end_station_name': 'station_name',
    'end_lat': 'station_lat',
    'end_lng': 'station_lng'
})

# Reset the index to have consecutive integer indices
end_stations_df = end_stations_df.reset_index(drop=True)

################################################################

# Perform an outer merge on station_id to find stations unique to either "start" or "end"
unique_stations_df = pd.merge(start_stations_df, end_stations_df, on='station_id', how='outer', indicator=True)

# Filter for rows where one of the station IDs is missing
unique_start_stations_df = unique_stations_df[unique_stations_df['_merge'] == 'left_only']
unique_end_stations_df = unique_stations_df[unique_stations_df['_merge'] == 'right_only']

# Drop the '_merge' column
unique_start_stations_df = unique_start_stations_df.drop('_merge', axis=1)
unique_end_stations_df = unique_end_stations_df.drop('_merge', axis=1)

# Display the unique start stations
print("Unique Start Stations:")
print(unique_start_stations_df)

# Display the unique end stations
print("\nUnique End Stations:")
print(unique_end_stations_df)
```

San Jose Depot is only used once and EB Test station is only used twice (and has fake coordinates) so we can forget these observations and only use the station in "start" for clustering

```python
# Select the relevant columns for stations
station_columns = ['start_station_id', 'start_station_name', 'start_lat', 'start_lng']

# Drop duplicates and keep the first occurrence
stations_df = df[station_columns].drop_duplicates(subset=['start_station_id'])

# Rename the columns for consistency
stations_df = stations_df.rename(columns={
    'start_station_id': 'station_id',
    'start_station_name': 'station_name',
    'start_lat': 'station_lat',
    'start_lng': 'station_lng'
})

# Reset the index to have consecutive integer indices
stations_df = stations_df.reset_index(drop=True)

# Display the resulting DataFrame
stations_df
```

### Let's first visualize where are the stations on the San Franciso map

```python
# pip install folium

import folium

# Assuming you have the 'stations_df' DataFrame with columns 'station_lat', 'station_lng', and 'station_name'

# Create a map centered around San Francisco
m = folium.Map(location=[37.7749, -122.4194], zoom_start=12)

# Add markers for each station
for index, row in stations_df.iterrows():
    folium.Marker(
        location=[row['station_lat'], row['station_lng']],
        popup=row['station_name']
    ).add_to(m)

# Display the map
m.save('bike_stations_map.html')
```

We can see there is 3 clear clusters

### Try 3 clusters

```python
# for 3 clusters

import folium
from sklearn.cluster import KMeans
import pandas as pd

# Assuming you have the 'stations_df' DataFrame with columns 'station_lat', 'station_lng', and 'station_name'

# Perform K-Means clustering with 3 clusters
n_clusters = 3
kmeans = KMeans(n_clusters=n_clusters, random_state=0).fit(stations_df[['station_lat', 'station_lng']])

# Add cluster labels to the DataFrame
stations_df['cluster'] = kmeans.labels_

# Create a map centered around San Francisco
m = folium.Map(location=[37.7749, -122.4194], zoom_start=12)

# Define a list of colors for clusters
cluster_colors = ['red', 'blue', 'green']

# Add markers for each station with cluster-specific colors
for index, row in stations_df.iterrows():
    cluster_color = cluster_colors[row['cluster']]
    folium.CircleMarker(
        location=[row['station_lat'], row['station_lng']],
        radius=5,
        color=cluster_color,
        fill=True,
        fill_color=cluster_color,
        fill_opacity=0.7,
        popup=row['station_name']
    ).add_to(m)

# Display the map
m.save('clustered_bike_stations_map.html')
```

### Try 20 clusters

```python
import folium
from sklearn.cluster import KMeans
import pandas as pd
import random

# Assuming you have the 'stations_df' DataFrame with columns 'station_lat', 'station_lng', and 'station_name'

# Perform K-Means clustering with 20 clusters
n_clusters = 20
kmeans = KMeans(n_clusters=n_clusters, random_state=0).fit(stations_df[['station_lat', 'station_lng']])

# Add cluster labels to the DataFrame
stations_df['cluster'] = kmeans.labels_

# Create a map centered around San Francisco
m = folium.Map(location=[37.7749, -122.4194], zoom_start=12)

# Generate random colors for clusters
cluster_colors = ['#{:02x}{:02x}{:02x}'.format(random.randint(0, 255), random.randint(0, 255), random.randint(0, 255) ) for _ in range(n_clusters)]

# Add markers for each station with cluster-specific colors
for index, row in stations_df.iterrows():
    cluster_color = cluster_colors[row['cluster']]
    folium.CircleMarker(
        location=[row['station_lat'], row['station_lng']],
        radius=5,
        color=cluster_color,
        fill=True,
        fill_color=cluster_color,
        fill_opacity=0.7,
        popup=row['station_name']
    ).add_to(m)

# Display the map
m.save('clustered_bike_stations_map_20_clusters.html')
```

```python
# Merge the cluster labels for start stations
df = df.merge(stations_df[['station_id', 'cluster']], left_on='start_station_id', right_on='station_id', how='left')

# Rename the 'cluster' column to 'start_station_cluster'
df = df.rename(columns={'cluster': 'start_station_cluster'})

# Merge the cluster labels for end stations
df = df.merge(stations_df[['station_id', 'cluster']], left_on='end_station_id', right_on='station_id', how='left')

# Rename the 'cluster' column to 'end_station_cluster'
df = df.rename(columns={'cluster': 'end_station_cluster'})
```

```python
# Drop rows with NA values in the specified columns
df = df.dropna(subset=['start_station_cluster', 'end_station_cluster'])

# Cast the new columns to integer type
df['start_station_cluster'] = df['start_station_cluster'].astype(int)
df['end_station_cluster'] = df['end_station_cluster'].astype(int)

# Drop the specified columns
df = df.drop(['station_id_x', 'station_id_y'], axis=1)
```
